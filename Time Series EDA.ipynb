## Resampling

```python
# Year End Frequency
df_tesla.resample(rule='YE').max()['Open'].plot()
```

```python
# Quarterly Start Frequency
# https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78/
df_tesla.resample(rule='QS').max()['Open'].plot()
```

```python
df_tesla.resample(rule='QS').max()
```

```python
# Business Year End Frequency
df_tesla.resample('BYE').max()
```

```python
df_tesla.resample('BYE').max().plot()
```

```python
# Business Quarterly Start Frequency
df_tesla.resample('BQS').max()
```

```python
df_tesla.resample('BQS').max().plot()
```

```python
df_tesla['Open'].resample(rule='YE').max()
```

```python
df_tesla['Open'].resample(rule='YE').mean()
```

```python
df_tesla['Open'].resample(rule='YE').mean().plot(kind='bar')
```

```python
# Rolling - Look at past N days and compute something. Example - Moving 5-day average

df_tesla['High'].rolling(window=5).mean().head(10)
```

```python
df_tesla['High'].rolling(window=5).mean().head(10)
```

```python
df_tesla['Open: 30 days rolling average'] = df_tesla['Open'].rolling(30).mean()
```

```python
df_tesla
```

```python
df_tesla[['Open','Open: 30 days rolling average']].plot(figsize=(15,6))
```

```python
df_microsoft = pdr.data.DataReader('MSFT','stooq', start='2016-01-01')
```

```python
df_microsoft = df_microsoft.sort_index()
```

```python
df_microsoft.index
```

```python
df_microsoft.loc['2017-01-01':'2023-01-01']
```

```python
df_microsoft.loc['2017-01-01':'2023-01-01']['High'].max()
```

```python
max_price = df_microsoft.loc['2017-01-01':'2023-01-01']['High'].max()
date_of_max_price = df_microsoft.loc['2017-01-01':'2023-01-01']['High'].idxmax()
```

```python
max_price
```

```python
date_of_max_price
```

```python
df_microsoft.loc['2021-11-01':'2021-11-30']['High']
```

```python
min_price = df_microsoft.loc['2017-01-01':'2023-01-01']['Low'].min()
date_of_min_price = df_microsoft.loc['2017-01-01':'2023-01-01']['Low'].idxmin()
```
